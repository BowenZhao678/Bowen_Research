# **ER**

```python
# please replace the following path with your path
df = pd.read_csv("/content/drive/MyDrive/Econ424/Final_csv/Bowen_Data.csv")
# state fixed effect
df['alpha'] = pd.Categorical(df['State'])
# year fixed effect
df['Tau'] = pd.Categorical(df['Year'])
# Set Indexes
df = df.set_index(['State', "Year"])
# Convert 'UMALE' to numeric
df['UMALE'] = df['UMALE'].str.rstrip('%').astype('float') / 100.0
# import model
from linearmodels.panel import PanelOLS

# Define and fit the model
m = PanelOLS(np.log(df.ER),
             exog=df[['WAGE', 'UMALE', 'POV']],
             entity_effects = True,
             time_effects = True)
m.fit(cov_type='clustered', cluster_entity=True)
```

# **UR**

```python
# please replace the following path with your path
df = pd.read_csv("/content/drive/MyDrive/Econ424/Final_csv/Bowen_Data.csv")
# state fixed effect
df['alpha'] = pd.Categorical(df['State'])
# year fixed effect
df['Tau'] = pd.Categorical(df['Year'])
# Set Indexes
df = df.set_index(['State', "Year"])
# Convert 'UMALE' to numeric
df['UMALE'] = df['UMALE'].str.rstrip('%').astype('float') / 100.0
# Convert 'UR' to numeric
df['UR'] = df['UR'].str.rstrip('%').astype('float') / 100.0

# import model
from linearmodels.panel import PanelOLS

# Define and fit the model
m = PanelOLS(df.UR,
             exog=df[['WAGE', 'UMALE', "POV"]],
             entity_effects = True,
             time_effects = True)
m.fit(cov_type='clustered', cluster_entity=True)
```
